```python
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
micro_org=microevent.dropna(subset=['org_name'], axis=0)
micro_org
```

```python
micro_org.columns
```

```python
micro_comp=micro_org[['hadm_id', 'chartdate', 'charttime', 'spec_type_desc','org_name']]
```

```python
micro_comp.columns=['hadm_id', 'mi_chartdate', 'mi_charttime', 'spec_type_desc','org_name']
```

```python
micro_comp
```

```python
micro_pu=micro_comp.drop_duplicates(subset=['mi_chartdate','org_name'], keep='first', inplace=False, ignore_index=False)
micro_pu
```

```python
micro_pu=micro_pu.dropna(subset=['hadm_id'], axis=0)
micro_pu
```

```python
icu_ab_join=pd.read_csv('/content/drive/MyDrive/mimiciv preprocessing/icu_ab_join.csv')
```

```python
icu_ab_join
```

```python
ab_micro_join = pd.merge(icu_ab_join, micro_pu, left_on='hadm_id', right_on='hadm_id', how='left')
```

```python
ab_micro_join
```

```python
ab_micro_join.info()
```

```python
ab_micro_join=ab_micro_join.dropna(subset=['org_name'], axis=0)
```
